# Simple MonteCarlo Simulations

This notebook explores very simple MonteCarlo simulations. <br>
It's purely a playground to try out the concepts behind MonteCarlo, to grow a better understanding of it and some practical hands-on experience. 


```python
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
```

## Simulating Dice Rolls

This first example will simulate the process of throwing 2 dices. 

First, let's consider our random variable to be the sum of the values of the two dices. <br>
Let's plot the probability distribution of that random variable. 


```python
def roll_dice(n): 
    """Rolls n times a dice and returns the series of resulting numbers.
    """
    return [random.randint(1, 6) for x in range(n)]
```

Let's plot the distribution of the result of throwing a dice a large number of times. <br>
Obviously we're expecting to see that **all numbers appear approximately the same number of times**, since they are generated from a **Uniform Distribution**. 

```python
plt.figure(figsize=(20, 3))

dice_results = roll_dice(10000)
sns.countplot(x=dice_results)
plt.show()
```

Now let's generate two dice rolls and sum the results. <br>
Let's **plot the distribution** of the results. <br>
What we expect to see is that 7 will have a higher probability, 6 and 8 a lower one, 5 and 9 lower even, etc. 

```python
def roll_two_dices_and_sum(n):
    """This method will roll two dices n number of times.
    """
    dice1 = np.array(roll_dice(n))
    dice2 = np.array(roll_dice(n))

    return dice1 + dice2
```

We will plot three plots: 
 - one based on 100 trials
 - one based on 1000 trials
 - one based on 10000 trials

That will help show that over a large number of trials, the distribution matches the expected distribution. 


```python
plt.figure(figsize=(20, 4))

plt.subplot2grid((1, 3), (0,0))
dice_results = roll_two_dices_and_sum(100)
sns.countplot(x=dice_results)
plt.title("100 Trials")

plt.subplot2grid((1, 3), (0,1))
dice_results = roll_two_dices_and_sum(1000)
sns.countplot(x=dice_results)
plt.title("1000 Trials")

plt.subplot2grid((1, 3), (0,2))
dice_results = roll_two_dices_and_sum(10000)
sns.countplot(x=dice_results)
plt.title("10000 Trials")

plt.show()
```
